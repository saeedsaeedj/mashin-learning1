```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import PolynomialFeatures 
```

```python
model=LinearRegression(normalize=True)
x=np.random.rand(200,1)
y=x**4+x**3+x**2+x
y=y+np.random.normal(scale=1,size=y.shape)
print(x.shape ,y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(x_test.shape)
print(x_train.shape)
```

```python
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
plt.plot(x_test,y_test, 'co')
plt.plot(x_train, y_train, 'bo')

x_line = np.arange(-0.25, 1.25, 0.01).reshape(-1,1)
y_line=model.predict(x_line)
plt.plot(x_line,y_line, 'r--')

print('Score of train set = {}\nScore of test set = {}' .format(model.score(x_train,y_train),model.score(x_test,y_test)))
```

```python

pol_features = PolynomialFeatures(degree = 2)  
x_pol_train = pol_features.fit_transform(x_train)
model.fit(x_pol_train,y_train)
x_pol_test=pol_features.fit_transform(x_test)
y_pol_predict=model.predict(x_pol_test)
plt.plot(x_test,y_test, 'co')
plt.plot(x_train, y_train, 'bo')

x_line = np.arange(-0.25, 1.25, 0.01).reshape(-1,1)
y_line=model.predict(pol_features.fit_transform(x_line))
plt.plot(x_line,y_line,'r-')
```

```python
l=[1,2,4,8,16]
for i in l:
    pol_features = PolynomialFeatures(degree = i)  
    x_pol_train = pol_features.fit_transform(x_train)
    model.fit(x_pol_train,y_train)
    x_pol_test=pol_features.fit_transform(x_test)
    y_pol_predict=model.predict(x_pol_test)
    plt.figure(i)
    plt.title(i)
    plt.plot(x_test,y_test, 'co')
    plt.plot(x_train, y_train, 'bo')

    print('Score of train set {} = {}\nScore of test set {} = {}' .format(i,model.score(x_pol_train,y_train),i,model.score(x_pol_test,y_test)))
    x_line = np.arange(-0.25, 1.25, 0.01).reshape(-1,1)
    y_line=model.predict(pol_features.fit_transform(x_line))
    plt.plot(x_line,y_line,'r-')
    
    
```
